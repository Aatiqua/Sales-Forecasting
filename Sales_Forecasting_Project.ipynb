# Sales Forecasting Project

Predict future sales using historical sales data with trend and seasonality.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

## Generate Sample Sales Data

```
dates = pd.date_range('2023-01-01', periods=730, freq='D')
trend = np.linspace(1000, 3000, len(dates))
seasonality = 250*np.sin(np.arange(len(dates))*2*np.pi/30)
noise = np.random.normal(0,100,len(dates))
sales = trend + seasonality + noise
df = pd.DataFrame({'Date':dates,'Sales':sales})
df.head()
```

## Train-Test Split

```
train = df.iloc[:-30]
test = df.iloc[-30:]
train.shape, test.shape
```

## Forecast Using 30-Day Moving Average

```
forecast_value = train['Sales'].tail(30).mean()
test = test.copy()
test['Forecast'] = forecast_value
test.head()
```

## Evaluate Model

```
mae = np.mean(np.abs(test['Sales']-test['Forecast']))
print('MAE:', round(mae,2))
```

## Visualization

```
plt.figure(figsize=(12,5))
plt.plot(train['Date'],train['Sales'],label='Train Sales')
plt.plot(test['Date'],test['Sales'],label='Actual Sales')
plt.plot(test['Date'],test['Forecast'],label='Forecast Sales')
plt.title('Sales Forecasting')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```
